```python
pdb = app.PDBFile("../PDBs/N439K/17309_CLONE0_GEN38_N439K_TER.pdb")
```

```python
for chain in pdb.topology.chains():
    print(chain.id, chain.index)
    for residue in chain.residues():
        print(residue.id, residue, residue.chain.id, residue.chain.index)
    print("\n")
```

```python
# Rename chains
for chain in pdb.topology.chains():
    if chain.index == 0:
        chain.id = 'E'
        chain.index = 2
    elif chain.index >= 1 and chain.index <=10:
        chain.id = 'X'
        chain.index = 4
    elif chain.index == 11:
        chain.id = 'A'
        chain.index = 0
    elif chain.index >= 12 and chain.index < 69:
        chain.id = 'X'
        chain.index = 3
    elif chain.index == 69:
        chain.id = 'A'
        chain.index = 1
```

```python
# Get first residues of each chain
d_first_residues = {}
for residue in pdb.topology.residues():
    if "A" not in d_first_residues and residue.chain.id == "A":
        d_first_residues["A"] = residue.id
    elif "E" not in d_first_residues and residue.chain.id == 'E':
        d_first_residues["E"] = residue.id
    elif "X" not in d_first_residues and residue.chain.id == 'X':
        d_first_residues["X"] = residue.id
        
# Rename residues
current_chain = ""
d_current_start = {"A": 18, "E": 332, "X": 729}
for residue in pdb.topology.residues():
    if residue.chain.id != current_chain: # If the residue's chain does not match the current_chain 
        current_chain = residue.chain.id 
    if residue.id != d_first_residues[current_chain]: # If it's the first residue of the chain
        d_current_start[current_chain] += 1
    # Update the residue id
    old_id = residue.id
    residue.id = str(d_current_start[current_chain])
```

```python
# Check that residue ids are changed in the bonds -- they are, but residue indexes are not
for bond in pdb.topology.bonds():
#     residue_A = bond[0].residue
#     residue_B = bond[1].residue
#     residue_A.id = d_old_to_new[residue_A.id]
#     residue_B.id = d_old_to_new[residue_B.id]
    print(bond, bond[0].residue.id, bond[0].residue.index, bond[1].residue.id)
```

```python
for chain in pdb.topology.chains():
    print(chain.index, chain.id)
    for residue in chain.residues():
        print(chain.id, residue.id, residue)
    print("\n")
```

```python
app.PDBFile.writeFile(pdb.topology, pdb.positions, open("test_rename.pdb", "w"), keepIds=True)
```

```python
# Get lists of chains to delete for each 
rbd_to_delete = []
ace2_to_delete = []
for chain in pdb.topology.chains():
    if chain.index in [0, 1, 3]:
        rbd_to_delete.append(chain)
    else:
        ace2_to_delete.append(chain)
```

```python
modeller_rbd = app.Modeller(pdb.topology, pdb.positions)
modeller_rbd.delete(rbd_to_delete)
```

```python
app.PDBFile.writeFile(modeller_rbd.topology, modeller_rbd.positions, open("test_rbd.pdb", "w"), keepIds=True)
```

```python
modeller_ace2 = app.Modeller(pdb.topology, pdb.positions)
modeller_ace2.delete(ace2_to_delete)
```

```python
app.PDBFile.writeFile(modeller_ace2.topology, modeller_ace2.positions, open("test_ace2.pdb", "w"), keepIds=True)
```
